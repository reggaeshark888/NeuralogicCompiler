### L2 regularization graph

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
lambda_reg = 1  # Regularization coefficient
weights = np.linspace(-2, 2, 400)  # Range of weight values from -2 to 2

# L2 Regularization term calculation
l2_regularization = lambda_reg * weights**2

# Plotting
plt.figure(figsize=(8, 4))
plt.plot(weights, l2_regularization, label='L2 Regularization Term')
plt.title('L2 Regularization Term for a Weight')
plt.xlabel('Weight (w)')
plt.ylabel('Regularization Term (w)')
plt.legend()
plt.grid(True)

plt.savefig('L2_Regularization_Plot.png', dpi=300, bbox_inches='tight')  # Save as PNG with high resolution

plt.show()
```

### Modified L2 regularization

```python
import numpy as np
import matplotlib.pyplot as plt

def J(w, alpha):
    if w >= alpha:
        return (w - 10)**2
    elif alpha > w > -alpha:
        return 0
    else:  # w <= -alpha
        return (w + 15)**2

# Define alpha
alpha = 5

# Generate a range of w values
w_values = np.linspace(-20, 20, 400)  # 400 points between -20 and 20

# Compute the J values for each w
J_values = np.array([J(w, alpha) for w in w_values])

# Plotting the function
plt.figure(figsize=(8, 4))
plt.plot(w_values, J_values, label=f'Piecewise function J(w), alpha = {alpha}')
plt.title('Graph of the Piecewise Function J(w)')
plt.xlabel('w')
plt.ylabel('J(w)')
plt.grid(True)
plt.legend()
plt.show()
```
